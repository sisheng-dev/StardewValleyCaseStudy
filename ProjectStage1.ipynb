## <font color="pink">Problem Formulation/Introduction</font>




Stardew Valley is a farming simulation game where players can cultivate crops across different seasons to earn profit. Each crop has unique attributes such as growth time, regrowth cycles, selling price, and seed cost. A key gameplay strategy involves deciding whether to grow single-harvest crops (harvested once per planting) or multi-harvest crops (which regrow multiple times without replanting).

This is why understanding which type of crop offers higher profit efficiency is valuable for optimizing farm management and for studying in-game economic balance, similar to how agricultural economists analyze ROI in real farming scenarios.


### Problem Formulation



> We aim to analyze whether multi-harvest crops yield higher average profit per day compared to single-harvest crops, based on in-game data.



### Research Question:



> Do multi-harvest crops in Stardew Valley yield significantly higher average profit per day than single-harvest crops?



### Null Hypothesis (H₀)



> There is no significant difference in average profit per day between multi-harvest and single-harvest crops.



### Alternative Hypothesis (H₁)



> Multi-harvest crops yield significantly higher average profit per day than single-harvest crops.



### Data Population



The population of interest is all cultivable crops in Stardew Valley, including their characteristics (growth time, regrowth, cost, profit, season, etc.) as implemented in the game’s most recent version.
Our dataset represents a snapshot of this population as compiled by the Kaggle contributor juletopi, who extracted in-game crop data from the Stardew Valley Wiki and related resources.

The dataset we will be using contains detailed information about all crops in Stardew Valley, including their name, growth and regrowth durations, seed cost, selling price, and harvest type. Each row represents a single crop species. The colums that are included in this dataset consist of:

**Crop** – name of the crop

**Growth Time** – days until first harvest

**Regrowth** – days between subsequent harvests (0 if single-harvest)

**Seed Price** – cost to purchase seeds

**Sell Price** – price per crop

**Harvests** – number of harvests per season

**Profit** – base profit per unit or per season

### Data Wrangling

**Structure** - 	The dataset is rectangular, where each row = crop. Columns include categorical (season, type) and numerical (price, days, profit).

**Granularity**	- Each record corresponds to a single crop type, not individual player harvest events which means that this dataset is an aggregate-level dataset.

**Scope** -	The scope of the dataset includes all crops available in Stardew Valley (since the version 1.6 update). This dataset covers all four seasons but excludes modded (game mods) or wild crops.

**Temporality** -	The data is static, not time-varying which reflects one version snapshot. Temporal dimension appears indirectly (growth time, regrowth).

**Faithfulness** - The data was manually collected from the official Stardew Valley Wiki, cross-checked by community contributors. It’s considered accurate but may omit hidden mechanics (e.g., quality multipliers or profession bonuses).

### Representativeness & Assumptions

The dataset is representative of all base-game crops in Stardew Valley, which fully covers our population of interest.
Our assumptions regarding the data include:

1. In-game prices and mechanics are constant (no profession modifiers, fertilizer, or skill boosts).

2. All crops are grown optimally within their respective seasons.

3. Seed cost and selling price represent standard market values (Pierre’s shop, base sell price).

### Sampling Method Guess

The data likely represents complete enumeration (census) rather than sampling. The author compiled all known crop entries from official game data. Therefore no random or stratified sampling occurred.
